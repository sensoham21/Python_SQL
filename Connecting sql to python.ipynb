```python
import pyodbc
```

```python
class Database_connect():
    
    def __init__(self, Driver1, Server1, Database1, Trusted_connection1):
        self.Driver1 = Driver1
        self.Server1 = Server1
        self.Database1 = Database1
        self.Trusted_connection1 = Trusted_connection1
        
    def connect_db(self):
        connection = pyodbc.connect(Driver= self.Driver1,
                            Server = self.Server1,
                            Database = self.Database1,
                            Trusted_connection = self.Trusted_connection1)
        return connection
    
    def execute_query(self,connection, query):
        cursor = connection.cursor()
        cursor.execute(query)
        return cursor
        
        
    def read(self):
        table = input("Please enter a valid table name: ")
        columns = input("Enter comma separated columns or enter * to view all: ")
        filter_criteria = input("Enter filter criteria: ")
        if filter_criteria == " " or filter_criteria == "":
            query = "Select {} FROM {}".format(columns,table)
        else:
            query = "Select {} FROM {} WHERE {}".format(columns,table,filter_criteria)
        connection = self.connect_db()
        cursor = self.execute_query(connection,query)
        for row in cursor:
            print(row)
        print()
        
    def update(self):
        table = input("Please enter a valid table name: ")
        columns = input("Enter column name: ")
        expression2 = input("Enter the new value: ")
        columns_1 = input("Enter second criteria column name: ")
        expression1 = input("Enter the criteria value: ")
        query = "UPDATE {} SET {} = {} WHERE {} = {}".format(table,columns,expression2,columns_1,expression1)
        connection = self.connect_db()
        self.execute_query(connection,query)
        connection.commit()
        
    def delete(self):
        table = input("Enter the table from where you want to delete: ")
        filter_criteria = input('Enter the filter criteria: ')
        query = "DELETE FROM {} WHERE {}".format(table,filter_criteria)
        connection = self.connect_db()
        self.execute_query(connection,query)
        connection.commit()
        
    def insert(self):
        table = input("Please enter a valid table name: ")
        insert_criteria = input("Please enter the column names: ")
        value_criteria = input("Please enter the values to be inserted: ")
        query = "INSERT INTO {}({}) VALUES({})".format(table,insert_criteria,value_criteria)
        connection = self.connect_db()
        self.execute_query(connection,query)
        connection.commit()

        
        
        
        
```
